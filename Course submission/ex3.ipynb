```python
from ipywidgets.widgets import widget
import json
#@title 
import ipywidgets as widgets
with open('students.json', 'r') as file:
  contents = json.load(file)  
file.close()

text1=widgets.Text()
text2=widgets.Text()
text3=widgets.Text()
text4=widgets.Text()
text5=widgets.Text()
text6=widgets.Text()

names = [f"{element['name']} {element['last name']}" for element in contents["students"]]
dropdown = widgets.Dropdown(
    options=names,
    value=names[0],
    description='Student : ',
    disabled=False,
)


                
def display_student(selected_value):
    for element in contents["students"]:
      if element["name"] == selected_value[0] and element["last name"] == selected_value[1]:
          text1.value=element["name"]
          text2.value=element["last name"]
          text3.value=element["email"]
          text4.value=", ".join(str(x) for x in element["courses"])
          text5.value=element["link"]
          if(element.get("tv") is not None):
            text6.value=element["tv"]
          else:
            text6.value=""

def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        selected_value = change['new'].split(" ")
        display_student(selected_value)



              
              
                  
display_student(names[0].split(" "))

dropdown.observe(on_change)

# display dropdown widget
button = widgets.Button(description="Click Here To Add The text !")
output = widgets.Output()
def on_button_clicked(b):
  for element in contents["students"]:
    if element["name"] == text1.value and element["last name"] == text2.value:
      element["tv"] = text6.value
  # Display the message within the output widget.
  with output:
    with open('students.json', 'w') as outfile:
      json.dump(contents, outfile, indent=4)
    print ( "Added To The File")
button.on_click(on_button_clicked)


#add to the existing file
display(dropdown,text1,text2,text3,text4,text5,button,text6,output)
```
